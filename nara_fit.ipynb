### Função para validação de valores:

```python
def validar_valores(pergunta, tipo): # Função para não repetir a validação
    while True:
        try:
            valor = tipo(input(pergunta))
            if valor < 0:
                print("Digite um valor válido!")
                continue
            print("Valor salvo com sucesso")
            return valor
        except ValueError:
            print("Digite um dado válido!")
```

### Função para calcular o IMC e avaliar o status da pessoa

```python
def calcular_avaliar(peso , altura): # Calcula o IMC e avalia o status do cadastro automaticamente 
    imc = round(peso / (altura**2), 2) 
    if imc >= 30: # Se quiser depois adiciona os graus
        status = "Obeso"
    elif imc >= 25:
        status = "Sobrepeso"
    elif imc >= 18.5:
        status = "Peso Normal"
    else:
        status = "Abaixo do peso"
    cadastrados["imc"].append(imc)
    cadastrados["status"].append(status)
```

### Cadastro + dicionario

```python
cadastrados = {
    "nomes": [],
    "idades": [],
    "pesos": [],
    "alturas": [],
    "imc": [],
    "status": []
}
def cadastro():
    while True: # Caso o usuário queira adicionar sem parar
        perguntas = [
            "Digite uma idade: ",
            "Digite um peso: ",
            "Digite uma altura: "
        ]
        nome = input("Digite o nome desejado: ")
        cadastrados["nomes"].append(nome)
        for i in perguntas:
            if perguntas.index(i) == 2:
                altura = validar_valores(i, float)
                cadastrados["alturas"].append(altura)
            elif perguntas.index(i) == 1:
                peso = validar_valores(i, float)
                cadastrados["pesos"].append(peso)
            else:
                idade = validar_valores(i, int)
                cadastrados["idades"].append(idade)
        index = len(cadastrados["nomes"]) - 1
        calcular_avaliar(cadastrados["pesos"][index], cadastrados["alturas"][index])
        print(f"Adicionado com sucesso!")
        if input("Deseja adicionar mais(S/N)?: ").lower() != "s": # Fazer tratamento de erros depois
            break
# Quando fizer a parte de nome faz um verificador, se o usuário digitar um nome que já existe ele pergunta se é outra pessoa ou a mesma. No caso se os valores forem muito parecidos.
```

### Função de listar + verificação para ver se a lista tem de fato alguém cadastrado

```python
def listar():
    if len(cadastrados["nomes"]) == 0:
        print("Sem usuários no sistema! Que tal cadastrar antes de checar a lista?")
    else:
        print(f"{"="*10} Lista de cadastrados {"="*10}")
        for i in range(len(cadastrados["nomes"])): # Por padrão vou usar o nomes 
            print(f"{i+1}° - {cadastrados["nomes"][i]} | {cadastrados["idades"][i]} anos | {cadastrados["pesos"][i]}Kg | {cadastrados["alturas"][i]}cm | IMC: {cadastrados["imc"][i]} | Status: {cadastrados["status"][i]}")
        print(f"{"="*42}")
```

### Estátisticas 

```python
def estatisticas():
    if len(cadastrados["nomes"]) == 0:
        print("Sem usuários no sistema! Que tal cadastrar antes de checar a lista?")
    else:
        media = sum(cadastrados["idades"]) / len(cadastrados["idades"])
        velho = max(cadastrados["idades"])
        novo = min(cadastrados["idades"])
        obesos = 0
        sobrepesos = 0
        normais = 0
        abaixos = 0
        for i in range(len(cadastrados["status"])):
            if cadastrados["status"][i] == "Obeso":
                obesos += 1
            elif cadastrados["status"][i] == "Sobrepeso":
                sobrepesos += 1
            elif cadastrados["status"][i] == "Peso Normal":
                normais += 1
            else:
                abaixos += 1
        print(f"{media}, {velho}, {novo}, {obesos}, {sobrepesos}, {normais}, {abaixos}")
```

### Menu Principal

```python
while True:
    print("1 - Deseja cadastrar um usuário?")
    print("2 - Consultar os usuários e valores?")
    print("3 - Consultar estáticas?")
    print("4 - Sair")
    try: 
        opcao = int(input("Escolha a opção: "))
    except ValueError:
        print("Digite um número")
    if opcao == 1:
        cadastro()
    elif opcao == 2:
        listar()
    elif opcao == 3:
        estatisticas()
    elif opcao == 4:
        if input("Deseja sair mesmo(S/N)?: ").lower() == "s":
            break
    else:
        print("Número inválido.")
```
